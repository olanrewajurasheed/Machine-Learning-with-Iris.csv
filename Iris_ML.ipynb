### Now we explore different supervised ML algorithms in sklearn
### We will be exploring 4 different classification approaches and calculate the accuracy in each case
#### 1. Decision Tree approach
#### 2. Logistic approach
#### 3. KNN approach
#### 4. Support vector machine SVM

```python
#import the necessary modules
from sklearn.linear_model import LogisticRegression  #for logistic regression approaches
from sklearn.neighbors import KNeighborsClassifier  #for KNN classifier approach
from sklearn import svm #for support vector machine approach
from sklearn.tree import DecisionTreeClassifier #for decision tree approach
from sklearn.model_selection import train_test_split 
from sklearn import metrics
```

#### Split data into training and test data 

```python
train, test = train_test_split(Iris_data, test_size=0.3) #commonly 70% training data
#30 for test data
```

### Specify training and testing data

```python
train_X = train[['sepal_length', 'sepal_width',	'petal_length',	'petal_width']]
train_y =train.species

test_X = test [['sepal_length', 'sepal_width',	'petal_length',	'petal_width']]
test_y =test.species
```

### 1) Decision Tree approach

```python
DT_model = DecisionTreeClassifier()
DT_model.fit(train_X, train_y)
DT_prediction = DT_model.predict(test_X)
DT_accuracy = metrics.accuracy_score(DT_prediction, test_y)
print ('The calculated accuracy of the Decision tree approach is', DT_accuracy)
```

### 2) Logistic approach 

```python
L_model = LogisticRegression()
L_model.fit(train_X, train_y)
L_prediction=L_model.predict(test_X)
L_accuracy = metrics.accuracy_score(L_prediction, test_y)
print ('The calculated accuracy of the Logistic approach is', L_accuracy)
```

### 3) KNN approach

```python
KNN_model = KNeighborsClassifier(n_neighbors=5)
KNN_model.fit(train_X, train_y)
KNN_prediction =KNN_model.predict(test_X)
KNN_accuracy =metrics.accuracy_score(KNN_prediction, test_y)
print ('The calculated accuracy of the KNN approach is', KNN_accuracy )
```

### 4) SVM approach

```python
svm_model=svm.SVC()
svm_model.fit(train_X, train_y)
svm_prediction = svm_model.predict(test_X)
svm_accuracy = metrics.accuracy_score(svm_prediction, test_y)
print ('The calculated accuracy of the SVM is', svm_accuracy)
```

### Seems the KNN appraoch gave the highest accuracy
